```python
import random
from matplotlib import pyplot as plt
from collections import defaultdict
```

```python
edges = [[1, 2, 3], [0, 4, 5], [0, 4, 6], [0, 5, 6], [1, 2, 7], [1, 3, 7], [2, 3, 7], [4, 5, 6]]

def transition(current_pos):
    choice = random.randrange(3)
    return edges[current_pos][choice]
```

```python
steps = []
for i in range(100000):
    current_vertex = transition(0)
    total_steps = 1
    while current_vertex != 0:
        current_vertex = transition(current_vertex)
        total_steps += 1
    steps.append(total_steps)
```

```python
plt.hist(steps, bins=len(set(steps)))
```

```python
sum(steps)/len(steps)
```

```python
running_average = [sum(steps[:i])/i for i in range(1,len(steps))]
```

```python
plt.plot(running_average[:500])
```

```python
steps = []
for i in range(10000000):
    current_vertex = transition(0)
    total_steps = 1
    while current_vertex != 0:
        current_vertex = transition(current_vertex)
        total_steps += 1
    steps.append(total_steps)
```

```python
count = defaultdict(int)

for i in steps:
    count[i] += 1
```

```python
count_avgs = {x:count_x/10000000.0 for (x,count_x) in count.items()}
```

```python
props = sorted([[a,b] for (a,b) in count_avgs.items()], key=lambda a:a[0])
```

```python
props_nums, props_props = [x[0] for x in props], [x[1] for x in props]
```

```python
plt.plot(props_nums, props_props)
```

```python
plt.plot(props_nums[1:], props_props[1:])
```
